```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=(100000,))
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(norm_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 5, 0, 1.1])  # Adjust x and y limits
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python

df = pd.read_csv(r"D:\Y4S1\DL\DL_Lab_IT22366290\DL_Practical\Au_nanoparticle_dataset.csv")

new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python

print(new_df.head(20))


print("Mean:\n", new_df.mean())
print("Std Dev:\n", new_df.std())
print("Quartiles:\n", new_df.quantile([0.25, 0.5, 0.75]))
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16,4))
for i, col in enumerate(new_df.columns):
    axes[i].hist(new_df[col], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)   # bivariate histogram
g.map_diag(sns.histplot, kde=True)  # hist + KDE
g.map_lower(sns.kdeplot, fill=True) # bivariate KDE
plt.show()
```
